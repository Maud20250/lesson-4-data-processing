#Task 3: Handling Null Values

```python
# Task 3: Handling Null Values
# Create a DataFrame with at least one null value in each column, then fill them
try:
    data_with_nulls = {
        'Name': ['Eve', 'Frank', np.nan, 'Grace'],
        'Age': [22, np.nan, 31, 27],
        'City': ['Toronto', 'Ottawa', 'Montreal', np.nan]
    }
    df_nulls = pd.DataFrame(data_with_nulls)
    print("Original DataFrame with null values:")
    print(df_nulls)
    print("\nNull value count per column:")
    print(df_nulls.isnull().sum())

    # Method 1: fillna() with a dictionary so each column gets an appropriate fill value
    df_filled = df_nulls.fillna({
        'Name': 'Unknown',
        'Age': 0,
        'City': 'Unknown'
    })
    print("\nDataFrame after filling nulls with fillna() (column-specific values):")
    print(df_filled)

    # Method 2: filling numeric nulls with the column mean (common alternative approach)
    df_mean_fill = df_nulls.copy()
    df_mean_fill['Age'] = df_mean_fill['Age'].fillna(df_mean_fill['Age'].mean())
    print("\nAlternative: Age column nulls filled with the column mean:")
    print(df_mean_fill[['Name', 'Age']])

    # Method 3: dropna() to remove any row containing a null value (shown for comparison)
    df_dropped = df_nulls.dropna()
    print("\nAlternative: rows with any null value dropped (dropna()):")
    print(df_dropped)

except Exception as e:
    print(f"An error occurred while handling null values: {e}")
```

#Task 5: Concatenation and Merging

```python
# Task 5: Concatenation and Merging
# Concatenate df1 and df2 vertically, then merge the result with df3 horizontally
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})
try:
    # Vertical concatenation: stacks df1 on top of df2 (same columns A, B)
    df_concat = pd.concat([df1, df2], axis=0, ignore_index=True)
    print("df1 and df2 concatenated vertically:")
    print(df_concat)
    print(f"Shape after concatenation: {df_concat.shape}")
    # Horizontal merge: df3 only has 2 rows, df_concat has 4 rows.
    # Since there's no common key column, we merge on the index and only
    # as many rows as df3 has will align without introducing NaNs.
    # To merge all 4 rows safely, we reset df3's index to align positionally
    # for the first 2 rows, then concatenate horizontally for the rest.
    df_merged = pd.merge(df_concat, df3, left_index=True, right_index=True, how='left')
    print("\nResult of df_concat merged with df3 horizontally (merge on index):")
    print(df_merged)
    print(f"Shape after merge: {df_merged.shape}")
    print("\nNote: df3 only has 2 rows while df_concat has 4 rows, so rows 2 and 3")
    print("receive NaN for columns C and D since there is no matching index entry in df3.")
except Exception as e:
    print(f"An error occurred during concatenation/merging: {e}")
```

#Task 6: Tuples and Sets

```python
# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers, then try to add an element to each
try:
    fruits_tuple = ('apple', 'banana', 'mango')
    numbers_set = {1, 2, 3, 4, 5}
    print(f"Original tuple: {fruits_tuple}")
    print(f"Original set: {numbers_set}")
    # Sets are mutable, so add() works directly
    numbers_set.add(6)
    print(f"\nSet after adding 6: {numbers_set}")
    # Tuples are immutable, so attempting to add an element raises an AttributeError
    try:
        fruits_tuple.add('orange') # this will fail; tuples have no add() method
    except AttributeError as e:
        print(f"\nError when attempting to add to the tuple: {e}")
    # Workaround: a "new" tuple can be created by concatenation, but the
    # original tuple object itself is never modified
    fruits_tuple_extended = fruits_tuple + ('orange',)
    print(f"Workaround (creates a NEW tuple instead): {fruits_tuple_extended}")
    print(f"Original tuple is unchanged: {fruits_tuple}")
    print("\nExplanation:")
    print("- Sets are mutable: elements can be added/removed in place with add()/remove().")
    print("- Tuples are immutable: once created, their contents cannot be changed.")
    print(" Any 'addition' requires building an entirely new tuple via concatenation.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

#Task 8: Functions and Lambda

```python
# Task 8: Functions and Lambda
# Define a regular function and an equivalent lambda function to square a number
def square(n):
    """Return the square of a number. Raises TypeError for non-numeric input."""
    if not isinstance(n, (int, float)):
        raise TypeError(f"Expected a number, got {type(n).__name__}")
    return n ** 2
# Lambda function performing the same operation
square_lambda = lambda n: n ** 2
try:
    test_inputs = [4, 7.5]
    for value in test_inputs:
        print(f"square({value}) using regular function = {square(value)}")
        print(f"square({value}) using lambda function = {square_lambda(value)}")
    print()
    # Demonstrate error handling with an invalid input
    try:
        square("not a number")
    except TypeError as e:
        print(f"Error handled gracefully: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

#Task 9: Iterators and Generators

```python
# Task 9: Iterators and Generators
# Create an iterator class and a generator function, both producing the first 5 even numbers
class EvenNumberIterator:
    """Custom iterator that yields the first `count` even numbers."""
    def __init__(self, count=5):
        self.count = count
        self.current = 0
        self.produced = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.produced >= self.count:
            raise StopIteration
        value = self.current
        self.current += 2
        self.produced += 1
        return value

def even_number_generator(count=5):
    """Generator function that yields the first `count` even numbers."""
    n = 0
    produced = 0
    while produced < count:
        yield n
        n += 2
        produced += 1

try:
    print("Using the custom iterator class:")
    even_iterator = EvenNumberIterator(5)
    iterator_results = list(even_iterator)
    print(iterator_results)

    print("\nUsing the generator function:")
    generator_results = list(even_number_generator(5))
    print(generator_results)

    # Confirm both approaches produce the same result
    assert iterator_results == generator_results, "Results differ between iterator and generator!"
    print("\nBoth approaches produced identical results:", iterator_results == generator_results)

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

#Task 10: Map, Reduce, and Filter

```python
# Task 10: Map, Reduce, and Filter
# Square all numbers (map), find the product of all numbers (reduce),
# and get only the even numbers (filter)
numbers = [1, 2, 3, 4, 5]
try:
    # map(): apply a squaring function to every element
    squared_numbers = list(map(lambda x: x ** 2, numbers))
    print(f"Original list: {numbers}")
    print(f"Squared numbers (map): {squared_numbers}")
    # reduce(): cumulatively multiply all numbers together
    product_of_numbers = reduce(lambda x, y: x * y, numbers)
    print(f"Product of all numbers (reduce): {product_of_numbers}")
    # filter(): keep only even numbers
    even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
    print(f"Even numbers only (filter): {even_numbers}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

#Task 11: Object-Oriented Programming

```python
import pandas as pd
import numpy as np
from functools import reduce

# Task 11: Object-Oriented Programming - Creating a Class
# Create a Rectangle class with area and perimeter methods
class Rectangle:
    """Represents a rectangle defined by length and width."""
    def __init__(self, length, width):
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive numbers.")
        self.length = length
        self.width = width

    def area(self):
        """Return the area of the rectangle."""
        return self.length * self.width

    def perimeter(self):
        """Return the perimeter of the rectangle."""
        return 2 * (self.length + self.width)

    def __repr__(self):
        return f"Rectangle(length={self.length}, width={self.width})"

try:
    rect1 = Rectangle(5, 3)
    rect2 = Rectangle(10, 4.5)

    for rect in (rect1, rect2):
        print(f"{rect}: area = {rect.area()}, perimeter = {rect.perimeter()}")

    # Demonstrate error handling for invalid dimensions
    try:
        invalid_rect = Rectangle(-2, 4)
    except ValueError as e:
        print(f"\nError handled gracefully when creating an invalid Rectangle: {e}")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
